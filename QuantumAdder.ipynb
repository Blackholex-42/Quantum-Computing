```python
from qiskit import *

qr = QuantumRegister(4)
cr = ClassicalRegister(2)

circuit = QuantumCircuit(qr,cr)
circuit.draw()


x=str(input("Enter Type of circuit"))
if "01" in x:
 circuit.x(1)

elif "10" in x:
 circuit.x(0)

elif "11" in x:

  circuit.x(0)
  circuit.x(1)


circuit.cx(0,2)
circuit.cx(1,2)
circuit.ccx(0,1,3)
circuit.measure(2,1)
circuit.measure(3,0)

circuit.draw()
```

```python
simulator = Aer.get_backend('qasm_simulator')

result = execute(circuit, backend=simulator).result()

from qiskit.visualization import plot_histogram

plot_histogram(result.get_counts(circuit))
```
